```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(
    "../data/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv",
    skiprows=4
)
```

```python
# Get the 10 countries with the highest population
top10 = df.nlargest(10, "2023")
```

```python
plt.figure(figsize=(12,6))

plt.bar(top10['Country Name'], top10['2023'])

plt.title("Top 10 Most Populated Countries (2023)", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)

plt.xticks(rotation=45)

plt.tight_layout()

plt.savefig("../images/population_bar_chart.png")

plt.show()
```

## Observations

China and India have the highest populations.
The population decreases gradually across the remaining countries.
The bar chart shows the distribution of population among the top 10 countries.

```python
plt.figure(figsize=(8,8))

top10.set_index('Country Name')['2023'].plot(kind='pie', autopct='%1.1f%%')

plt.ylabel("")
plt.title("Population Share of Top 10 Countries (2023)")

plt.show()
```
